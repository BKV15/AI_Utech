    <h1 dir='rtl'>سوال 1 - R<sup>2<sup></h1>
    <p dir='rtl' style='text-align:justify'>
        R<sup>2</sup> مقداری برای تطابق میزان واریانس متغیر وابسته توسط متغیرهای مستقل می‌باشد. درواقع R<sup>2</sup> همان R معمولی است. R میزان تطابق دو متغیر را نسبت به هم می سنجد و مقدار آن بین -1 تا 1 می‌باشد که مقادر نزدیک به -1 و 1 نشان دهنده ارتباط قوی بین دو متغیر است و مقادیر نزدیک 0 نشان دهنده وجود نداشتن رابطه بین دو متغیر را نشان می‌دهد. 
    </p>
    <p dir='rtl' style='text-align:justify'>
    R<sup>2</sup> بدلیل راحت‌تر بودن تفسیر و محاسبه آن معمولا کاربرد بیشتری دارد. برای محاسبه R<sup>2</sup> می‌توان از فرمول R<sup>2</sup> = ( Var(mean) – Var(fit) ) / Var(mean) استفاده کرد که عدد آن بین 0 تا 1 می‌باشد و می‌توان به صورت درصد هم آن را تفسیر کرد.
    </p>
    <p dir='rtl' style='text-align:justify'>
     تفسیر R<sup>2</sup>  به صورت میزان درصدی از واریانس متغیر وابسته که با در نظرگرفتن متغیرهای مستقل قابل توضیح است، گفته می‌شود. هرچه این مقدار به 100 نزدیک تر باشد، معادله خط عبور داده شده از دیتاها هم دقیق تر است. البته برای اطلاع از اینکه آیا مقدار محاسبه شده از نظر آماری درست و قابل اعتماد هست باید از p-value استفاده کرد.
    </p>
    <p dir='rtl' style='text-align:justify'>
    همچنین مقدار R<sup>2</sup> برای چندین متغیر مستقل که ارتباط خطی مشخصی با متغیر وابسته ندارند کمتر از مقدار آن برای یک متغیر مستقل با ارتباط خطی مناسب نخواهد بود. زیرا ظرایب متغیرهای این چنینی در هنگام محاسبه least squares نزدیک به صفر خواهد بود.
    </p>
    <p dir='rtl' style='text-align:justify'>
    از R<sup>2</sup> برای مشخص کردن جهت ارتباط دو متغیر نمی‌توان استفاده کرد، برای مثال اگر دو متغیر ارتباط خطی قوی داشته باشند نمی‌توان گفت آیا این ارتباط به صورت منفی است یا مثبت.
    </p>
</div>

    <h1 dir='rtl'>سوال 2 - ROC</h1>
    <p dir='rtl' style='text-align:justify'>
 منحنی ROC میزان کارایی مدل Classification را می‌سنجد. این منحنی که خود یک منحنی احتمال است، به وسیله رسم True Positive Rate (TPR) بر False Positive Rate (FPR) در آستانه های مختلف به دست می‌آید. TPR = Recall = Sensitivity = TP / TP + FN می باشد و FPR = 1 – Specificity = FP / TN + FP می باشد. 
    </p>
    <p dir='rtl' style='text-align:justify'>
 مساحت زیر این نمودار به AUC معروف است. این مقدار در واقع میزان جدایش مدل ساخته شده را مشخص می‌کند به این صورت که مقادیر نزدیک به 1 نشان دهنده میزان جداپذیری و تشخیص بالا می‌باشد. برای مثال مدل به درستی مقادیر Y را Y و مقادیر N را N تشخیص می‌دهد. در صورتی که این مقدار به 0.5 برسد، مدل ما تقریبا هیچ توانایی در تمایز Y از N ندارد و احتمال تشخیص هرکدام به صورت 50 درصد می‌باشد. و درصورتی که این مقدار به 0 برسد، یعنی مدل ما کاملا برعکس تشخصی خواهد داد. به صورتی که مقادیر Y را N و مقادیر N را Y تشخیص خواهد داد.
    </p>
    <p dir='rtl' style='text-align:justify'>
 در صورتی که بیشتر از دو حالت برای تشخیص داشته باشیم (Multi-Classification) باید از روش One vs All استفاده کرد. برای مثال در صورتی که سه کلاس a , b , c داشته باشیم باید در هر مرحله نمودار ROC برای a در برابر b و c ، b در برابر a و c و در آخر c در برابر a و b را رسم کنیم
    </p>
</div>
